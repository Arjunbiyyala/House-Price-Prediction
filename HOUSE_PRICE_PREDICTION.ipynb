```python
x=df[['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup','Latitude','Longitude']]
y=df['price']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=90)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=90)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
pred=lm.predict(x_test)
```

```python
pred.shape
```

```python
pred[:5]
```

```python
y_test[:5]
```

```python
from sklearn import metrics
print("MAE : ",metrics.mean_absolute_error(y_test,pred))
print("MSE : ",metrics.mean_squared_error(y_test,pred))
print("RSME : ",np.sqrt(metrics.mean_absolute_error(y_test,pred)))
print("RMSE : ",np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,pred)
print(f"R2 score {score:.4f}")
```

```python
print("cofficent :",lm.coef_)
```

```python
y_test
```

```python
df1=pd.DataFrame()
```

```python
df1['actual price']=y_test
df1['predicted price']=pred
```

```python
df1['Error']=y_test-pred
```

```python
df1=df1.reset_index(drop=True)
```

```python
df1
```

```python
min_price = min(df1['actual price'].min(),
                df1['predicted price'].min())

max_price = max(df1['actual price'].max(),
                df1['predicted price'].max())

plt.scatter(df1['actual price'],
            df1['predicted price'],
            alpha=0.4)

plt.plot([min_price, max_price],
         [min_price, max_price],
         'r--',
         linewidth=2)

plt.title('Actual vs Predicted Prices with Reference Line')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.show()
```

```python
plt.scatter(df1['predicted price'],df1['Error'])
plt.ylabel('Error')
plt.xlabel('Predicted Price')
plt.title('residual plot')
```
